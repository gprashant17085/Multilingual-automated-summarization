```
def load_data(filename):
    file = open(filename, encoding='utf-8')
    text = file.read()
    file.close()
    text = text.strip().split('\n')
    return text
```

```
pred = load_data("pred.txt")
golden = load_data("data/Giga/task1_ref0.txt")
rouge = Rouge()

count = 0
rouge1_f = 0.0
rouge1_p = 0.0
rouge1_r = 0.0
rouge2_f = 0.0
rouge2_p = 0.0
rouge2_r = 0.0
rougel_f = 0.0
rougel_p = 0.0
rougel_r = 0.0

for p, g in zip(pred, golden):
    try:
      scores = rouge.get_scores(p, g)
      rouge1_f += scores[0]['rouge-1']['f']
      rouge2_f += scores[0]['rouge-2']['f']
      rougel_f += scores[0]['rouge-l']['f']
      rouge1_p += scores[0]['rouge-1']['p']
      rouge2_p += scores[0]['rouge-2']['p']
      rougel_p += scores[0]['rouge-l']['p']
      rouge1_r += scores[0]['rouge-1']['r']
      rouge2_r += scores[0]['rouge-2']['r']
      rougel_r += scores[0]['rouge-l']['r']
      count += 1
    except:
      print(p)
      print(g)

print(rouge1_f / count)
print(rouge1_p / count)
print(rouge1_r / count)
print(rouge2_f / count)
print(rouge2_p / count)
print(rouge2_r / count)
print(rougel_f / count)
print(rougel_p / count)
print(rougel_r / count)
```
