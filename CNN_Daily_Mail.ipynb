```
import pandas as pd
import numpy as np

def open_file(filename):
    with open(filename, encoding='utf-8') as file:
      text = file.read()
    return text.strip().split('\n')

def get_data(name):
  src_path = name+'.txt.src'
  data = open_file(src_path)
  df = pd.DataFrame(data,columns=['source'])
  target_path = name+'.txt.tgt.tagged'
  target = open_file(target_path)
  df['target'] = pd.DataFrame(target)
  print(f"{name} Data Size: {df.shape}")
  df['len_source']= df['source'].apply(lambda x:len(x))
  df['len_target']= df['target'].apply(lambda x:len(x))
  print(df.head())
  return df

train = get_data('train')
test = get_data('test')
val = get_data('val')
```
